```python
from translate import Translator
translator= Translator(from_lang="chinese",to_lang="english")
datas=[]
op=0
with open(r'data\relations_3.csv','r') as f:
    
    
    for line in f:
        op+=1
        if op%10==0:
            print(op)
        try:
            row=line.rstrip('\n').split(',')
            row[1]=int(row[1])
            translation = translator.translate(row[0]).split(' ')
            # print(translation)
            string=''
            string_count=0
            for i in translation:
                string_count+=1
                if string_count>1:
                    string+=i.capitalize() 
                else:
                    string+=i.lower() 
            row.append(string)
            datas.append(row)
        except:
            pass
```
